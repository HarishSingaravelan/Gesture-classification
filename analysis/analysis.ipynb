```python
import pandas as pd
```

```python
fileNames=[]
for i in range(1,10):
    fileName=f"../Lab_1/left{i}.csv"
    fileNames.append(fileName)
numeric_df = pd.read_csv(fileNames[0]).select_dtypes(include='number')
print(" ".join([col for col in numeric_df.columns]))

for filename in fileNames:
    df = pd.read_csv(filename)
    df.columns = df.columns.str.strip()
    # print(df["Arm"].count())

    numeric_df = df.select_dtypes(include='number')
    for col in numeric_df.columns:
        print(f" {round(numeric_df[col].median(),4)}", end="|")
    print("\n")
```

```python
fileNames=[]
for i in range(1,10):
    fileName=f"../Lab_1/fisk{i}.csv"
    fileNames.append(fileName)
numeric_df = pd.read_csv(fileNames[0]).select_dtypes(include='number')
print(" ".join([col for col in numeric_df.columns]))

for filename in fileNames:
    df = pd.read_csv(filename)
    df.columns = df.columns.str.strip()
    # print(df["Arm"].count())

    numeric_df = df.select_dtypes(include='number')
    for col in numeric_df.columns:
        print(f" {round(numeric_df[col].median(),4)}", end="|")
    print("\n")

```

```python
# List to store median values
median_data = []

# Generate file names
fileNames = [f"../Lab_1/left{i}.csv" for i in range(1, 10)]

# Process each file
for filename in fileNames:
    df = pd.read_csv(filename)
    df.columns = df.columns.str.strip()
    
    # Select only numeric columns
    numeric_df = df.select_dtypes(include='number')
    
    # Compute median and store as a dictionary
    median_values = round(numeric_df.median(),5).to_dict()
    median_values["label"] = "left"  # Add filename for reference
    
    # Append to list
    median_data.append(median_values)

# Convert list of dictionaries to DataFrame
left_median_df = pd.DataFrame(median_data)

# Save to CSV (optional)
left_median_df.to_csv("left_median_values.csv", index=False)

# Display the resulting DataFrame
print(left_median_df)
```

```python
# List to store median values
median_data = []

# Generate file names
fileNames = [f"../Lab_1/fist{i}.csv" for i in range(1, 10)]

# Process each file
for filename in fileNames:
    df = pd.read_csv(filename)
    df.columns = df.columns.str.strip()
    
    # Select only numeric columns
    numeric_df = df.select_dtypes(include='number')
    
    # Compute median and store as a dictionary
    median_values = round(numeric_df.median(),5).to_dict()
    median_values["label"] = "fist"  # Add filename for reference
    
    # Append to list
    median_data.append(median_values)

# Convert list of dictionaries to DataFrame
fist_median_df = pd.DataFrame(median_data)

# Save to CSV (optional)
fist_median_df.to_csv("fist_median_values.csv", index=False)

# Display the resulting DataFrame
print(fist_median_df)
```

```python
combined_df = pd.concat([fist_median_df, left_median_df], axis=0, ignore_index=True)
combined_df.to_csv("combined_df.csv")
```
